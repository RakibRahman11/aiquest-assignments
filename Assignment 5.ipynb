```python
import pandas as pd
import numpy as np
```

```python
data = {'Price' : [110, 105, 115, 120, 110, 130, 150, 100, 105]}
df = pd.DataFrame(data)
df
```

Normalization

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df['Price_norm'] = scaler.fit_transform(df[['Price']])
df.head()
```

Standardization

```python
from sklearn.preprocessing import StandardScaler

std = StandardScaler()
df['Price_stand'] = std.fit_transform(df[['Price']])
df.head()
```

Log Transformation

```python
from sklearn.preprocessing import FunctionTransformer

df['Price_log'] = FunctionTransformer(np.log1p).fit_transform(df[['Price']])
df.head()
```

Robust Scaler

```python
from sklearn.preprocessing import RobustScaler

df['Price_robust'] = RobustScaler().fit_transform(df[['Price']])
df.head()
```

Max Absolute Scaler

```python
from sklearn.preprocessing import MaxAbsScaler

scaler = MaxAbsScaler()
df['Price_maxScaler'] = scaler.fit_transform(df[['Price']])
df.head()
```
